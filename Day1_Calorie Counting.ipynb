```python
import pandas as pd
```

```python
# fill in the file path
path ="***File Path***"

df = pd.read_csv(path,header=None)
```

#### The original file was modified by search and replace all empty cells in the column into 0

```python
df.head(10)
```

```python
L = df[0].tolist()
```

```python
L
```

```python
cal_carry = []
cal = 0
for i in range(len(L)):
    
    if L[i] != 0:
        cal = L[i] + cal
        
        
    if L[i] == 0:
        cal_carry.append(cal)
        #print(cal_carry)
        cal = 0
        
index = cal_carry.index(max(cal_carry))
                        
print("The maximum calorie carried is:", max(cal_carry), "The Elf carrying the most caloreis is:", index+1)
```

```python
carry_sorted = sorted(cal_carry, reverse=True)
```

```python
print ("Total calorie carried by top three is", 
       carry_sorted[0]+carry_sorted[1]+carry_sorted[2])

print("Their numbers are:", cal_carry.index(carry_sorted[0])+1,
      cal_carry.index(carry_sorted[1])+1,
      cal_carry.index(carry_sorted[2])+1)
```
